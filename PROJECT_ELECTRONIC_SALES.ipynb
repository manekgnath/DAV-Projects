```python
# Importing the library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# reading the Excel file data  & Display contents
df = pd.read_excel('/content/Electronics sales report.xlsx')
df
```

```python
# Retrieve the dimensions of the DataFrame (rows, columns)
df.shape
```

```python
# Display a concise summary of the DataFrame, including column names, data types, non-null counts, and memory usage.
df.info()
```

```python
df.duplicated().sum()
```

```python
df['Items Purchased'] = df['Items Purchased'].fillna(method='bfill')

```

```python
df['Warranty Extension'] = df['Warranty Extension'].fillna(method='bfill')
```

```python
df['Gender'] = df['Gender'].fillna(method='bfill')
```

```python
df['Payment Method'] = df['Payment Method'].fillna(method='ffill')
```

```python
df['Region'] = df['Region'].fillna(method='ffill')
```

```python
df['Age'] = df['Age'].fillna(method='ffill')
```

```python
df['Membership Status'] = df['Membership Status'].fillna(method='ffill')
```

```python
df['Loyalty Score'] = df['Loyalty Score'].fillna(method='ffill')
```

```python
df['Preferred Visit Time'] = df['Preferred Visit Time'].fillna(method='ffill')
```

```python
df['Product Category'] = df['Product Category'].fillna(df['Product Category'].mode()[0])
```

Count plot

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(data=df, x="Product Category", hue="Gender", palette="coolwarm", ax=ax)
ax.set_title("Count of Product Categories by Gender")
ax.set_xlabel("Product Category")
ax.set_ylabel("Count")
#save_plot(fig, "count_plot_category_by_gender.png")
```

piechart

```python
fig, ax = plt.subplots(figsize=(8, 6))
time_count = df["Preferred Visit Time"].value_counts()
time_count.plot(kind='pie', autopct='%1.1f%%', ax=ax, wedgeprops={'width': 0.3}, cmap="Blues")
ax.set_title("Preferred Visit Time Distribution (Donut Chart)")
#save_plot(fig, "donut_chart_preferred_time.png")
```

stacked barchart

```python
fig, ax = plt.subplots(figsize=(8, 6))
gender_time_count = df.groupby(["Preferred Visit Time", "Gender"]).size().unstack()
gender_time_count.plot(kind='bar', stacked=True, ax=ax, cmap="Set2")
ax.set_title("Count of Products by Gender and Preferred Visit Time")
ax.set_xlabel("Preferred Visit Time")
ax.set_ylabel("Count")
fig.savefig("stacked_bar_gender_time.png", dpi=300)
#save_plot(fig, "stacked_bar_gender_time.png")
```

 Heatmap: Correlation Between Rating, Revenue, and Age


Explanation: A heatmap showing the correlation between Rating, Revenue, and Age.

```python
fig, ax = plt.subplots(figsize=(8, 6))
correlation_matrix = df[["Store Rating", "Revenue", "Age"]].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
ax.set_title("Correlation Heatmap: Store Rating, Revenue, and Age")
#save_plot(fig, "heatmap_Store Rating_revenue_age.png")
```

5. KDE Plot: Kernel Density Estimation of Rating



Explanation: A KDE plot showing the smoothed distribution of Rating.

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.kdeplot(data=df, x="Store Rating", fill=True, ax=ax, color="green")
ax.set_title("Kernel Density Estimation of Store Rating")
ax.set_xlabel("Store Rating")
ax.set_ylabel("Density")
fig.savefig("kde_plot_Store Rating.png", dpi=300)
#save_plot(fig, "kde_plot_Store Rating.png")
```
